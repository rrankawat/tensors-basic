```python
import torch
import matplotlib.pyplot as plt
```

```python
t_one = torch.tensor([1, 2, 3])
t_two = torch.tensor([1, 2, 3])

# Addition
print(t_one + t_two)

# Multiplication
print(t_one * t_two)
print(t_one * 5)

# Dot product
dot_product = torch.dot(t_one, t_two)
print(dot_product)
```
